```python
csv_file = open('cmt_filter\\0x2f50ab197f950e0c2184cf5d804f4141502cd987_0x.csv', 'r')
csv_reader = csv.reader(csv_file, delimiter=',')
gg,gt,sg,st = [],[],[],[]
for row in csv_reader:
    if(row[0] == 'S'):
        sg.append(int(row[1]))
        st.append(int(row[3]))
    elif(row[0] == 'G'):
        gg.append(int(row[1]))
        gt.append(int(row[3]))
gc,sc = [],[]
tmp = 0
count = 0
for item in range(len(gt)):
    count += 1
    if(gt[item] != tmp):
        if(tmp != 0):
            gc.append([gt[item],count])
        tmp = item
count = 0
tmp = 0
for item in range(len(st)):
    count += 1
    if(st[item] != tmp):
        if(tmp != 0):
            sc.append([st[item],count])
        tmp = item

fig, axs = plt.subplots(1, 1,figsize=(9,5))
axs.scatter(gt, gg, s=2, alpha=.2, color='red', label="Out of gas") 
axs.scatter(st, sg, s=2, alpha=.2, color='green', label="Success") 
axs.grid()
axs.legend()
axs.set_xlabel('Block Number')
axs.set_ylabel('Gas (Scatter)')

axs2 = axs.twinx()
axs2.step([i[0] for i in gc],[i[1] for i in gc], marker='', linewidth=1, where='post', color='#FF6666', label="Out of gas")
axs2.step([i[0] for i in sc],[i[1] for i in sc], marker='', linewidth=1, where='post', color='#32CD32', label="Success")
axs2.set_ylabel('Count (Step)')
axs2.legend()

csv_file.close()
```
